```python
top_courses = data.sort_values(by='num_subscribers',ascending=False).head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_courses['course_title'], top_courses['num_subscribers'], color='skyblue')
plt.title('Top Courses by Subscribers')
plt.xlabel('Course Title')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
#trend of course publications over time
```

```python
data['published_timestamp'] = pd.to_datetime(data.published_timestamp)
```

```python
course_counts_by_year = data['published_timestamp'].dt.year.value_counts()
```

```python
plt.figure(figsize=(12,6))
course_counts_by_year.plot(kind='bar',color='pink')

plt.title('Number of courses published each year')
plt.xlabel('Years')
plt.ylabel('Number of courses')
plt.xticks(rotation=45)
plt.show()
```
